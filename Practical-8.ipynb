# Practical No.8: Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', '2024-06-30'),
    'Open': np.random.randint(100, 200, 182)
})
df['Close'] = df['Open'] + np.random.randint(-10, 10, 182)
df.set_index('Date')[['Open', 'Close']].plot(figsize=(20,5), title='Opening vs Closing Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
