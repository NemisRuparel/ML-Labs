# Practical No.8: Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Create a Sample DataFrame ---
data = {
    'Date': pd.to_datetime([
        '2023-08-01', '2023-08-02', '2023-08-03', '2023-08-04', '2023-08-07',
        '2023-08-08', '2023-08-09', '2023-08-10', '2023-08-11', '2023-08-14',
        '2023-08-15', '2023-08-16', '2023-08-17', '2023-08-18', '2023-08-21'
    ]),
    'Open': [172.5, 173.1, 171.0, 170.5, 172.0, 175.3, 177.2, 176.8, 178.9, 180.1,
             179.5, 178.8, 177.0, 175.5, 176.2],
    'Close': [173.0, 172.8, 170.2, 171.8, 174.9, 176.5, 176.0, 178.5, 180.3, 179.8,
              178.6, 177.4, 175.8, 176.9, 177.8]
}
df = pd.DataFrame(data)

# Set the 'Date' column as the index for easier slicing
df.set_index('Date', inplace=True)

# --- 2. Specify the Date Range ---
start_date = '2023-08-04'
end_date = '2023-08-16'

# --- 3. Filter the DataFrame for the specified dates ---
filtered_df = df.loc[start_date:end_date]

# --- 4. Create the Line Plot ---
plt.figure(figsize=(10, 5))  # Set the figure size for better readability

# Plot the 'Open' and 'Close' price columns
plt.plot(filtered_df.index, filtered_df['Open'], label='Open Price')
plt.plot(filtered_df.index, filtered_df['Close'], label='Close Price')

# --- 5. Customize the Plot ---
plt.title(f'Stock Prices from {start_date} to {end_date}', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Stock Price', fontsize=12)
plt.legend()  # Display the legend
plt.grid(True)  # Add a grid for easier analysis
plt.xticks(rotation=45) # Rotate date labels for better visibility
plt.tight_layout() # Adjust layout to make room for rotated date labels

# --- 6. Show the Plot ---
plt.show()
```
